```python
# import libraries

import requests
from bs4 import BeautifulSoup as bs
```

```python
# Sending http request to get the html content

link = "https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft"
page_response = requests.get(link)
print(page_response)

print("Successful")
```

```python
# Parsing the html content for data extraction

soup = bs(page_response.content,"html.parser")
```

```python
# Analyze the html content and extract the useful data

Aircrafts = soup.find('table', class_ = "wikitable")
Aircrafts
```

```python
# th = table headers

columns = Aircrafts.find_all('th')
columns
```

```python
# Extracting the headers

headers = []

for header in columns:
    headers.append(header.text.strip())
    
print(headers)
```

```python
# Function to remove text in square brackets using split as it is unwanted

def remove_brackets(text):
    return text.split('[')[0].strip()
```

```python
# Extracting the rows data and make sure that each row has 6 elements to get the correct tabular format

table_rows = Aircrafts.find_all('tr')

rows = []

for tr in table_rows:
    data = []
    for td in tr.find_all('td'):  # Extracting the table data
        data.append(remove_brackets(td.text).strip())
    if data and len(data)>=5:
        rows.append(data)
    
rows[0]
```

```python
# Convert the data into a DataFrame

import pandas as pd
df = pd.DataFrame(rows, columns=headers[:6])

df = df.replace('','Nil').replace('None','Nil')
df
```

```python
# Converting dataframe to csv

df.to_csv('Aircrafts.csv', index=False)
```
